## 🔄 3. Data Science Jarayoni: CRISP-DM

### 📊 CRISP-DM (Cross-Industry Standard Process for Data Mining)

```
    1. Biznes Tushunish
           ⬇️
    2. Ma'lumotlarni Tushunish 
           ⬇️
    3. Ma'lumotlarni Tayyorlash
           ⬇️
    4. Modellashtirish
           ⬇️
    5. Baholash
           ⬇️
    6. Deployment (Joriy qilish)
```

---

### 🎯 **1. Biznes Tushunish (Business Understanding)**

**Maqsad:** Loyiha maqsadlarini aniq belgilash

**Savollar:**
- Qanday muammo hal qilinmoqda?
- Muvaffaqiyat qanday o'lchanadi?
- Qanday resurslar mavjud?

**Misol:** *"Mijozlar nima uchun kompaniyani tark etishmoqda?"*

---

### 📊 **2. Ma'lumotlarni Tushunish (Data Understanding)**

**Maqsad:** Mavjud ma'lumotlarni o'rganish

**Ishlar:**
- Ma'lumotlar hajmi va sifatini tekshirish
- Etishmayotgan qiymatlarni aniqlash
- Dastlabki statistika hisobotlarini tayyorlash

**Misol:** *1000 ta mijoz, 50 ta xususiyat, 5% noma'lum ma'lumot*

---

### 🧹 **3. Ma'lumotlarni Tayyorlash (Data Preparation)**

**Maqsad:** Ma'lumotlarni modellashtirish uchun tayyorlash

**Jarayonlar:**
- Tozalash (cleaning)
- O'zgartirish (transformation) 
- Birlashtirish (integration)
- Tanlab olish (selection)

**Vaqt:** Odatda loyihaning 60-80% vaqti bu bosqichga ketadi!

---

### 🤖 **4. Modellashtirish (Modeling)**

**Maqsad:** Machine Learning modellarini yaratish

**Bosqichlar:**
1. Model turini tanlash
2. Parametrlarni sozlash
3. Modelni o'qitish
4. Natijalarni tekshirish

**Misol modellari:**
- 🌳 Decision Tree
- 🧠 Neural Network
- 📊 Linear Regression

---

### ✅ **5. Baholash (Evaluation)**

**Maqsad:** Model sifatini o'lchash

**Metrikalari:**
- Aniqlik (Accuracy)
- Precision va Recall
- F1-Score
- ROC-AUC

**Savol:** Model biznes maqsadlariga mos keladimi?

---

### 🚀 **6. Deployment (Joriy qilish)**

**Maqsad:** Modelni real muhitda ishlatish

**Yo'llar:**
- Web ilovasi
- Mobil dastur
- API service
- Avtomatik tizim

**Monitoring:** Model ishlashini doimiy kuzatish
