## ðŸ”‘ Individual Amaliyot uchun Yechim

Mana individual mashq uchun to'liq yechim:

```python
# Individual Amaliyot uchun Yechim
student_scores = [85, 92, 78, 96, 88, 74, 91, 83, 87, 95, 79, 89, 93, 82, 86, 90, 77, 94, 81, 88]
scores_series = pd.Series(student_scores)

print("YECHIM - Talaba Baholari Tahlili")
print("=" * 50)

# Markaziy Tendentsiya
mean_score = scores_series.mean()
median_score = scores_series.median()
mode_score = scores_series.mode()

print(f"Markaziy Tendentsiya:")
print(f"  O'rtacha: {mean_score:.2f}")
print(f"  Mediana: {median_score:.2f}")
print(f"  Moda: {mode_score.values[0] if len(mode_score) > 0 else 'Yagona moda yo\'q'}")

# Tarqalish
range_score = scores_series.max() - scores_series.min()
variance_score = scores_series.var()
std_score = scores_series.std()

print(f"\nTarqalish:")
print(f"  Diapaz: {range_score}")
print(f"  Dispersiya: {variance_score:.2f}")
print(f"  Standart Og'ish: {std_score:.2f}")

print(f"\nTalqin:")
print(f"  - O'rtacha ({mean_score:.2f}) va mediana ({median_score:.2f}) juda yaqin,")
print(f"    bu ma'lumotlar taxminan simmetrik va katta haddan tashqari qiymatlar yo'qligini ko'rsatadi.")
print(f"  - {std_score:.2f} standart og'ish ko'pchilik talabalar")
print(f"    o'rtachadan {std_score:.2f} ball farqda ({mean_score:.2f} ball) olganini anglatadi.")
print(f"  - Bu o'rtacha darajadagi tarqalishni ifodalaydi.")
```

## ðŸŒŸ Bonus Qo'shimcha Vazifa 

**O'rtacha o'quvchilar uchun:**

Quyidagi stsenariyni tahlil qiling:
- Kompaniya 2 ta mahsulot ishlab chiqaradi: A va B
- A mahsulot: [100, 120, 110, 130, 105, 125, 115, 135, 108, 122] (ming so'm)
- B mahsulot: [200, 180, 220, 160, 240, 190, 210, 170, 230, 200] (ming so'm)

**Savollar:**
1. Qaysi mahsulot ko'proq barqaror (kam o'zgaruvchan) narxga ega?
2. Qaysi mahsulotning o'rtacha narxi yuqori?
3. Variatsiya koeffitsientini hisoblang va taqqoslang (CV = std/mean * 100%)

```python
# Bonus Vazifa Yechimi
print("BONUS VAZIFA - Mahsulot Narxlari Taqqoslash")
print("=" * 55)

# Ma'lumotlar
product_a = [100, 120, 110, 130, 105, 125, 115, 135, 108, 122]
product_b = [200, 180, 220, 160, 240, 190, 210, 170, 230, 200]

a_series = pd.Series(product_a)
b_series = pd.Series(product_b)

# Asosiy statistikalar
print("MAHSULOT A (ming so'm):")
print(f"  O'rtacha: {a_series.mean():.1f}")
print(f"  Standart og'ish: {a_series.std():.2f}")

print("\nMAHSULOT B (ming so'm):")
print(f"  O'rtacha: {b_series.mean():.1f}")
print(f"  Standart og'ish: {b_series.std():.2f}")

# Variatsiya koeffitsienti
cv_a = (a_series.std() / a_series.mean()) * 100
cv_b = (b_series.std() / b_series.mean()) * 100

print(f"\nVARIATSIYA KOEFFITSIENTLARI:")
print(f"  Mahsulot A: {cv_a:.2f}%")
print(f"  Mahsulot B: {cv_b:.2f}%")

print(f"\nXULOSALAR:")
print(f"  1. Barqarorlik: {'A mahsulot' if cv_a < cv_b else 'B mahsulot'} ko'proq barqaror")
print(f"     (kichik variatsiya koeffitsienti)")
print(f"  2. O'rtacha narx: {'A mahsulot' if a_series.mean() > b_series.mean() else 'B mahsulot'} yuqori o'rtacha narxga ega")
print(f"  3. B mahsulot narxi yuqori lekin ko'proq o'zgaruvchan,")
print(f"     A mahsulot esa barqaror va bashorat qilish oson.")
```
