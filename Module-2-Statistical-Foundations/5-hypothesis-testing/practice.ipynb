```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
n = 10
trials = 10_000

heads = np.random.binomial(n, 0.9, size=trials)

"""
50 -> 4800 -> gerb
50 -> 5200 -> sonli

90 -> 9000 -> gerb
10 -> 1000 -> sonli
"""

# p-value ~ P(X >= 16 | fair coin)
p_emp = np.mean(heads >= 16)
print(f"Empirik p-qiymat (>=16 gerb): {p_emp:.4f}")

# Tarqatish grafigi
plt.figure()
_ = plt.hist(heads, bins=range(0, n+2), edgecolor='black', align='left')
plt.title("10 marta tashlashda gerb soni taqsimoti (10 000 simulyatsiya)")
plt.xlabel("Gerblar soni")
plt.ylabel("Chastota")
plt.show()
```

# Qaror qanday bo‘ladi?
	•	Biz odatda α = 0.05 (5%) ni “chegara” sifatida olamiz.
	•	bizda p = 0.0057, ya’ni 0.57% < 5%.
	•	Demak, “Bunday natija odatda bo‘lmaydi, juda g‘alati” → H₀ (adolatli tanga) ni rad qilamiz.
	•	Xulosa: Tangada adolatsizlik bo‘lishi mumkin.

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

np.random.seed(1)
n1, n2 = 30, 30
group_A = np.random.normal(loc=75, scale=10, size=n1)  # shokolad
group_B = np.random.normal(loc=70, scale=10, size=n2)  # shokoladsiz

t_stat, p_val = stats.ttest_ind(group_A, group_B, equal_var=False)
print(f"t-stat = {t_stat:.3f}, p-value = {p_val:.4f}")

plt.figure()
plt.hist(group_A, alpha=0.6, bins=12, edgecolor='black', label='A (shokolad)')
plt.hist(group_B, alpha=0.6, bins=12, edgecolor='black', label='B (yo‘q)')
plt.title("Ikki guruh ballari taqsimoti")
plt.xlabel("Ball")
plt.ylabel("Chastota")
plt.legend()
plt.show()
```
